```python
# criando um driver para utilizar este IP
from pynq import DefaultIP

class AddDriver(DefaultIP):
    def __init__(self, description):
        super().__init__(description=description)
    
    # definir este driver criado para o ip especifico
    bindto = ['user.org:user:adder_ip:1.0']
    
    # função para realizar a soma
    def add(self, values):
        # escreve no registrador AXI de entrada
        self.write(0x00, values)
        # Le o valor do registrador AXI de saida
        return self.read(0x04)
```

```python
# gravando novamente na placa
overlay = Overlay('/home/xilinx/adder_axilite.bit')
# pode verificar agora que o driver criado AddDriver foi associado ao IP adder_ip
overlay?
```
